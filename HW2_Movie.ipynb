```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
import re
import random
```

```python
base_url = 'https://movies.yahoo.com.tw/'
#mv1 = "https://movies.yahoo.com.tw/movieinfo_main/%E4%B8%80%E5%80%8B%E5%B7%A8%E6%98%9F%E7%9A%84%E8%AA%95%E7%94%9F-a-star-is-born-8511"
#mv2 = 'https://movies.yahoo.com.tw/movieinfo_main/%E8%AA%B0%E5%85%88%E6%84%9B%E4%B8%8A%E4%BB%96%E7%9A%84-dear-ex-8809'
his = ['movieinfo_main/%E8%AA%B0%E5%85%88%E6%84%9B%E4%B8%8A%E4%BB%96%E7%9A%84-dear-ex-8809']
url = base_url + his[-1]
url_list =[]
url_list.append(url)
#mv_list=[]

for i in range(20):
    mv_list=[]
    html = urlopen(url_list[-1]).read().decode("utf-8")
    soup = BeautifulSoup(html,features='lxml')
    
    print(str(i+1),soup.find('h1').get_text(),soup.find('h3').get_text())#,soup.find('h3'))
    descrip = soup.find('meta',{"name":"twitter:description","content":re.compile(r"(.*)")})['content']
    
    movie_cnname = soup.select_one(".movie_intro_info_r h1").text
    movie_egname = soup.select_one(".movie_intro_info_r h3").text
    movie_date = soup.select(".movie_intro_info_r span")[0].text
    movie_length = soup.select(".movie_intro_info_r span")[1].text
    movie_company = soup.select(".movie_intro_info_r span")[2].text
    movie_imdb = soup.select(".movie_intro_info_r span")[3].text
    movie_director = soup.select(".movie_intro_list")[0].text.strip()
    movie_actor = soup.select(".movie_intro_list")[1].text
    movie_plot = soup.select_one(".gray_infobox_inner span").text
#    aaa=[]
#    for i in soup.select(".level_name a"):
#        aaa.append(i.text.strip())
    mv_list.append(movie_cnname)
    mv_list.append(movie_egname)
#    mv_list.append(aaa)
    mv_list.append(movie_date)
    mv_list.append(movie_length)
    mv_list.append(movie_company)
    mv_list.append(movie_imdb)
    mv_list.append(movie_director)
    mv_list.append(movie_actor)
    mv_list.append(movie_plot)
    descrip = soup.find('meta',{"name":"twitter:description","content":re.compile(r"(.*)")})['content']
    '''
    print("movie_date:",movie_date)
    print("movie_length:",movie_length)
    print("movie_company:",movie_company)
    print("movie_director:",movie_director)
    print("movie_actor:",movie_actor)
    print("movie_plot:",movie_plot)
    print("description:",descrip)
    print(" MV URL:", url_list[-1])
    print("\n\n")
    '''
    next_mvs = soup.find_all('a',{'href':re.compile(r"https://movies.yahoo.com.tw/movieinfo_main/(%.{2})+")})
    cnt =0
    if (len(next_mvs)!=0):
        next_movie = random.sample(next_mvs,1)[0]['href']
        
        while(next_movie in url_list):
            next_movie = random.sample(next_mvs,1)[0]['href']
            cnt+=1
            if cnt >100:
                print("can't find new movie")
                break
            
        url_list.append(next_movie)
        
    else:
        url_list.pop()
    if(i==0):
        df = pd.DataFrame(data=[mv_list])#,
                        # columns = ['電影名稱(中文)', '電影名稱(英文)', '上映日期',
                        #            '片長','發行公司', 'IMDb分數', '導演', '演員', '劇情'])
    else:
        df_tmp=pd.DataFrame([mv_list])
        #print("!!",df_tmp)
        df = df.append(df_tmp,ignore_index=True,verify_integrity=True)
        #print(df)
    if cnt >100:
              #  print("can't find new movie")
                break
    
        
        
print(len(mv_list))
```
